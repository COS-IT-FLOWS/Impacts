```python
import pandas as pd
import geopandas as gpd

def convert_csv_to_shapefile(csv_file, shp_file):
    # Read the CSV file into a Pandas DataFrame
    df = pd.read_csv(csv_file)

    # Create a geometry column from the 'wkt' column
    # Applying the function to the entire 'Wkt' column to create a GeoSeries
    df['geometry'] = gpd.GeoSeries.from_wkt(df['Wkt'])
    # Creating a GeoDataFrame
    gdf = gpd.GeoDataFrame(df, geometry='geometry', crs='EPSG:4326')

    # Write the GeoDataFrame to a Shapefile
    gdf.to_file(shp_file)

    return None

# Example usage
convert_csv_to_shapefile("/content/drive/MyDrive/Lowerbasin/roads_lower.csv",
                         "/content/drive/MyDrive/Lowerbasin/roadsntwrk_lower.shp")
```

```python
import osgeo.ogr as ogr
import pandas as pd

# Load the shapefiles
roads_shapefile_path = '/content/drive/MyDrive/colab files/roadsntwrk.shp'
tidal_flood_shapefile_path = '/content/drive/MyDrive/colab files/tidal_flood_lsg_ekm.shp'

# Load the tidal flood shapefile
tidal_flood_data = ogr.Open(tidal_flood_shapefile_path)
if tidal_flood_data is None:
    print(f"Error: Could not open shapefile at '{tidal_flood_shapefile_path}'")
    # Handle the error (e.g., exit the script, raise an exception)
else:
    tidal_flood_layer = tidal_flood_data.GetLayer()

    # Get the field names from the tidal flood shapefile
    tidal_flood_layer_defn = tidal_flood_layer.GetLayerDefn()
    tidal_flood_field_names = [tidal_flood_layer_defn.GetFieldDefn(i).GetName() for i in range(tidal_flood_layer_defn.GetFieldCount())]
    print("Available field names in tidal flood shapefile:", tidal_flood_field_names)

    # Load the roads shapefile
    roads_data = ogr.Open(roads_shapefile_path)
    if roads_data is None:
        print(f"Error: Could not open shapefile at '{roads_shapefile_path}'")
        # Handle the error (e.g., exit the script, raise an exception)
    else:
        roads_layer = roads_data.GetLayer()

        # Create a list to store the results
        results = []

        # Iterate over each polygon in the tidal flood shapefile
        for tidal_flood_feature in tidal_flood_layer:
            tidal_flood_geom = tidal_flood_feature.GetGeometryRef()
            panchayat_name = tidal_flood_feature.GetField("NAME")  # Replace with your actual field name if exists in the field_names

            # Initialize the sum of linestring length
            sum_length = 0

            # Iterate over each feature in the roads shapefile
            for roads_feature in roads_layer:
                roads_geom = roads_feature.GetGeometryRef()

                # Check if the roads feature intersects with the tidal flood polygon
                if tidal_flood_geom.Intersects(roads_geom):
                    # Calculate the length of the linestring
                    length = roads_geom.Length()
                    # sum_length = 0

                    # Add the length to the sum
                    sum_length += length

            # Append results to list
            results.append({
                "Panchayat Name": panchayat_name,
                "Sum of Length": sum_length
            })

        # Create a Pandas DataFrame from results
        df = pd.DataFrame(results)

        # Print the table
        print(df)
```

```python
import geopandas as gpd

# Load the road shapefile
roads_gdf = gpd.read_file('/content/drive/MyDrive/colab files/roadsntwrk.shp')

# Load the boundary shapefile
boundary_gdf = gpd.read_file('/content/drive/MyDrive/colab files/tidal_flood_lsg_ekm.shp')

# Define a projected CRS
projected_crs = 'EPSG:4326'

# Create a table to store the results
results = []

# Loop through each polygon in the boundary shapefile
for i, polygon in boundary_gdf.iterrows():
    # Clip the road shapefile within the current polygon
    clipped_roads = gpd.clip(roads_gdf, polygon.geometry)

    # Re-project the clipped roads to the projected CRS
    clipped_roads_projected = clipped_roads.to_crs(projected_crs)

    # Calculate the sum of length of linestring within the polygon
    total_length = clipped_roads_projected.length.sum()

    # Append the results to the table
    results.append({
        'Polygon Name': polygon['name'],
        'Sum of Length': total_length
    })

# Print the table
print("Polygon Name\tSum of Length")
for result in results:
    print(f"{result['Polygon Name']}\t{result['Sum of Length']:.2f}")
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

def process_and_analyze_roads(thresholds, elevation_shp_path, road_shp_path, boundary_shp_path):
    for threshold in thresholds:
        output_shp = f"/content/drive/MyDrive/Lowerbasin/EXTRACTED_FILES/{threshold}poly_lower.shp"

        # Load the elevation shapefile (at or below threshold)
        elevation_shp = gpd.read_file(output_shp)

        # Load the road shapefile
        road_shp = gpd.read_file(road_shp_path)

        # Perform the intersection operation to clip the roads
        clipped_roads = gpd.overlay(road_shp, elevation_shp, how='intersection')

        # Save the clipped roads to a new shapefile
        clipped_roads.to_file('/content/drive/MyDrive/Lowerbasin/clipped_roads{threshold}.shp', driver='ESRI Shapefile')

        # Load the clipped roads shapefile
        clipped_roads = gpd.read_file('/content/drive/MyDrive/Lowerbasin/clipped_roads{threshold}.shp')

        # Load the boundary shapefile
        boundary_shp = gpd.read_file(boundary_shp_path)

        # Reproject the clipped roads and boundary shapefile to 'EPSG:32643'
        clipped_roads = clipped_roads.to_crs(epsg=32643)
        boundary_shp = boundary_shp.to_crs(epsg=32643)

        # Initialize a list to store results
        results = []

        # Iterate through each polygon in the boundary shapefile
        for index, polygon in boundary_shp.iterrows():
            # Filter the clipped roads that intersect with the current polygon
            roads_within_polygon = clipped_roads[clipped_roads.intersects(polygon.geometry)]

            # Sum the lengths of the filtered roads
            total_length = roads_within_polygon.length.sum()

            # Append the results to the list
            results.append({
                'Polygon Name': polygon['name'],  # Adjust if necessary
                'Sum of Length (km)': round(total_length / 1000,2)  # Convert to kilometers
            })

        # Create a summary DataFrame from the results
        summary_table = pd.DataFrame(results)

        # Print the summary table
        print(summary_table.to_string(index=False))

      #  # Plot the original elevation and clipped roads
      #   fig, ax = plt.subplots(figsize=(12, 8))
      #   elevation_shp.plot(ax=ax, color='lightblue', label='Elevation', alpha=0.5)
      #   clipped_roads.plot(ax=ax, edgecolor='orange', linewidth=2, label='Clipped Roads', alpha=0.7)
      #   boundary_shp.plot(ax=ax, edgecolor='red', linewidth=1, label='Boundary', alpha=0.5)

      #   # Set the title and labels
      #   ax.set_title(f"Clipped Roads and Elevation for Threshold {threshold}m")
      #   ax.set_xlabel("Longitude")
      #   ax.set_ylabel("Latitude")

      #   # Add a legend to the plot
      #   ax.legend()

      #   # Show the plot
      #   plt.show()
# Example usage
thresholds = [1]
elevation_shp_path = "/content/drive/MyDrive/Lowerbasin/EXTRACTED_FILES/{threshold}poly_lower.shp"
road_shp_path = '/content/drive/MyDrive/Lowerbasin/roadsntwrk_lower.shp'
boundary_shp_path = '/content/drive/MyDrive/Lowerbasin/lower_basin_lsg_panchayat.shp'

process_and_analyze_roads(thresholds, elevation_shp_path, road_shp_path, boundary_shp_path)
```

```python
import geopandas as gpd

# Load the road shapefile
roads_gdf = gpd.read_file('/content/drive/MyDrive/Lowerbasin/roadsntwrk_lower.shp')

# Load the boundary shapefile
boundary_gdf = gpd.read_file('/content/drive/MyDrive/Lowerbasin/lower_basin_lsg_panchayat.shp')

# Load the elevation threshold shapefile
elevation_gdf = gpd.read_file('/content/drive/MyDrive/Lowerbasin/EXTRACTED_FILES/15poly_lower.shp')  # Adjust path as necessary

# Define a projected CRS
projected_crs = 'EPSG:32643'

# Create a table to store the results
results = []

# Loop through each polygon in the boundary shapefile
for i, polygon in boundary_gdf.iterrows():
    # Clip the road shapefile within the current polygon
    clipped_roads = gpd.clip(roads_gdf, polygon.geometry)

    # Re-project the clipped roads to the projected CRS
    clipped_roads_projected = clipped_roads.to_crs(projected_crs)

    # Filter elevation geometries that intersect with the current polygon
    elevation_within_polygon = elevation_gdf[elevation_gdf.intersects(polygon.geometry)]

    # Initialize total length
    total_length = 0

    # Check if there are elevation features within the polygon
    if not elevation_within_polygon.empty:
        # Loop through each elevation geometry
        for _, elevation in elevation_within_polygon.iterrows():
            elevation_geom = elevation.geometry

            # Filter the clipped roads that intersect with the elevation geometry
            roads_within_elevation = clipped_roads_projected[clipped_roads_projected.intersects(elevation_geom)]

            # Sum the lengths of the filtered roads
            total_length += roads_within_elevation.length.sum()

    # Append the results to the table
    results.append({
        'Polygon Name': polygon['name'],  # Adjust if necessary
        'Sum of Length (km)': total_length / 1000  # Convert to kilometers
    })

# Print the results
print("Polygon Name\tSum of Length (km)")
for result in results:
    print(f"{result['Polygon Name']}\t{result['Sum of Length (km)']:.2f}")
```
